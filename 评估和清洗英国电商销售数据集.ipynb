# 分析目标

## 此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

# 简介

## 原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况。涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结场纪念品、圣诞礼品等等。该公司的客户群体生要包括批发商和个人消费者。其中批发商占据了相当大的比例。

## 数据每列的含义如下：
## InvoiceNo：发票号码。6位数,作为交易的通一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
## StockCode:产品代用。5位数。作为产品的唯一标识符。
## Description:产品名称。
## Quantity：产品在交易中的数量。
## InvoiceDate:发票日期和时间。交易发生的日期和时间。
## UnitPrice：单价。价格单位为英镑（£）。
## CustomerID:客户编号，5位数，作为客户的唯一标识符。
## Country:国家名称。客户所居住的国家的名称。

## 清理数据

需要进行的操作有：
- InvoiceDate的数据变量类型改成日期时间
- CustomerID的数据变量类型改为字符串
- Description变量缺失的观察值删除
- Quantity为负数的删掉
- UnitPrice为负数的删掉

```python
cleaned_data=original_data.copy()
cleaned_data.head()
```

```python
cleaned_data["InvoiceDate"]=pd.to_datetime(cleaned_data["InvoiceDate"],format='mixed')
cleaned_data["InvoiceDate"]
```

https://pandas.pydata.org/pdeps/0004-consistent-to-datetime-parsing.html
可以查找这个的用法

```python
cleaned_data["CustomerID"]=cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

```python
cleaned_data["CustomerID"]=cleaned_data["CustomerID"].str.slice(0,5)
cleaned_data["CustomerID"]
```

或者写slice(0,-2)，这里ID肯定不是有小数的所以后的要删掉

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
```

```python
cleaned_data["Description"].isnull().sum()
```

```python
cleaned_data=cleaned_data[cleaned_data["Quantity"]>=0]
```
